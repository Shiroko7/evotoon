```python
## IMPORTS 
import evotoon
from data_classes import CatParam, IntParam, FloatParam
```

```python
## MAKE SEED
SEED = evotoon.make_seed(39)
```

```python
## LHS EXAMPLE

float_params = [
	FloatParam("float1", 0.0, 1.0),
	FloatParam("float2", 20.0, 200.0),
]
int_params = [
	IntParam("int1", 0, 100),
	IntParam("int2", 0, 50),
]
cat_params = [
	CatParam("cat1", [str(i) for i in range(0, 1000, 3)]),
	CatParam("cat2", ["small", "medium", "big"]),
]

initial_values = evotoon.initialization(10, float_params, int_params, cat_params)

for value in initial_values:
	print(value)
```

```python
## EXAMPLE USE WITH NAIVE TUNNER USING ILSMKP

# SET PARAMETERS RANGE
int_params = [IntParam("k",1,50)]

initial_values = evotoon.initialization(
	poblation_size=10, int_parameters_list=int_params
)
print(initial_values)

# SET ENVIRONMENT FOR THE ILSMKP ALGORITHM TO TUNE
function_kwargs = {
	"executable_path": "./ILSMKP/ILSMKP",
	"instance_path": "./ILSMKP/instances/gk01.dat",
	"seed": SEED,
	"evaluations": 10000,
}

# CALL THE TUNER
output, val = evotoon.naive_tunning(
	evotoon.execute_ILSMKP, initial_values, **function_kwargs
)
print("BEST:", output, val)
```
