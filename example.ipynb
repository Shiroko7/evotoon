```python
## EXAMPLE USE WITH NAIVE TUNNER USING ILSMKP
separator = "".join(["-" for i in range (10)])
# SET PARAMETERS RANGE
int_params = [IntParam("k",1,150)]

initial_values = evotoon.initialization(
	poblation_size=20, int_parameters_list=int_params
)
print(separator, "Intial Values", separator)
print(initial_values)

# SET ENVIRONMENT FOR THE ILSMKP ALGORITHM TO TUNE
instance_list = ["./ILSMKP/instances/mknap1_5.txt", "./ILSMKP/instances/mknap1_6.txt", "./ILSMKP/instances/mknap1_7.txt"]
seed_list = [SEED + i for i in range(len(instance_list))]
print(separator, "Instances running and seeds", separator)
for ins,seed in zip(instance_list,seed_list):
	print("instance:", ins, "seed:", seed)

function_kwargs = {
	"executable_path": "./ILSMKP/ILSMKP",
	"instance_list": instance_list,
	"seed_list": seed_list,
	"evaluations": 10000,
}
print(separator, "Running The Tunner", separator)
# CALL THE TUNER
output, val = evotoon.naive_tunning(
	evotoon.execute_ILSMKP, initial_values, **function_kwargs
)
print(separator, "Results", separator)
print("BEST:", output, val)
```
