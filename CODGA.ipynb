```python
## IMPORTS 
import evotoon
from data_classes import CatParam, IntParam, FloatParam


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import seaborn as sns#

import random
```

```python
## MAKE SEED
SEED = evotoon.make_seed(283)
separator = "--------------------------------------------------------"
```

```python
## EXAMPLE WITH CodGA

# Env configuration
poblation_size = 10

# Parameter settings
float_params = [
	FloatParam("p_c", 0.0001, 0.1, 4),
	FloatParam("p_m", 0.0001, 0.1, 4),
]
int_params = [
	IntParam("N", 1, 100),
	IntParam("t_max", 1, 1000),
]
cat_params = [
	CatParam("c_0", [1, 2]),
]

all_params = float_params + int_params + cat_params
all_params = {param.name: param for param in all_params}

initial_batch = evotoon.initialization(poblation_size, float_params, int_params, cat_params)

# Tuner parameters

update_cycle = 5
budget = 30

# SET ENVIRONMENT FOR THE ACO ALGORITHM TO TUNE
instance_list = [
	"./InstanciasNKLandscapes/nk_2_38/nk_38_2.4",
	"./InstanciasNKLandscapes/nk_2_38/nk_38_2.9",
	"./InstanciasNKLandscapes/nk_2_38/nk_38_2.15",	
]
seed_list = [SEED + i for i in range(len(instance_list))]
print(separator, "Instances running and seeds", separator)
for ins,seed in zip(instance_list,seed_list):
	print("instance:", ins, "seed:", seed)

function_kwargs = {
	"executable_path": "./CodGA/ga-nk",
	"instance_list": instance_list,
	"seed_list": seed_list,
	"output": "xd.txt"
}

print(pd.DataFrame(initial_batch))
```

```python
evotoon.evo_tunning(all_params=all_params, budget=budget, poblation_size=poblation_size, update_cycle=update_cycle, initial_batch=initial_batch, execute_algorithm=evotoon.execute_CodGA, returning_type="ABSOLUTE_OPTIMAL_DIFF", **function_kwargs)
```
