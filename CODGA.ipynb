```python
## IMPORTS 
import evotoon
from data_classes import CatParam, IntParam, FloatParam


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import seaborn as sns#

import random
```

```python
## MAKE SEED
SEED = evotoon.make_seed(765)
separator = "--------------------------------------------------------"
```

```python
## EXAMPLE WITH CodGA

# Env configuration
poblation_size = 50

# Parameter settings
float_params = [
	FloatParam("p_c", 0.001, 1, 3),
	FloatParam("p_m", 0.001, 1, 3),
]
int_params = [
	IntParam("N", 1, 50),
]

all_params = float_params + int_params 
all_params = {param.name: param for param in all_params}



initial_batch = evotoon.initialization(poblation_size, float_params, int_params)

# Tuner parameters

update_cycle = 5
budget = 5000

# SET ENVIRONMENT FOR THE ACO ALGORITHM TO TUNE
path = "./InstanciasNKLandscapes/Instances-Training"
n = 10

instance_list = evotoon.choose_instances(path, n , SEED)

instance_list, seed_list = evotoon.select_seeds(instance_list, n, SEED)

print(separator, "Instances running and seeds", separator)
for ins,seed in zip(instance_list,seed_list):
	print("instance:", ins, "seed:", seed)

function_kwargs = {
	"executable_path": "./CodGA/ga-nk",
	"instance_list": instance_list,
	"seed_list": seed_list,
	"output": "xd.txt",
	"t_max": 1000 
}

print(pd.DataFrame(initial_batch))
```

```python
import warnings
warnings.filterwarnings("ignore")
evotoon.evo_tunning(all_params=all_params, float_params=float_params, int_params=int_params, budget=budget, poblation_size=poblation_size, initial_batch=initial_batch, execute_algorithm=evotoon.execute_CodGA, returning_type="ABSOLUTE_OPTIMAL_DIFF", **function_kwargs)
```
